#### Basic Linear System operations

```python
A = np.array([
        [-1, 3],
        [3, 2]
    ], dtype=np.dtype(float))

b = np.array([7, 1], dtype=np.dtype(float))

print("Matrix A:")
print(A)
print("\nArray b:")
print(b)
```

```python
x = np.linalg.solve(A, b)

print(f"Solution: {x}")
```

```python
print(np.linalg.det(A))
```

```python
B = np.array([[1,2,1], [2,1,1], [-1,2,1]])

print(np.linalg.det(B))
```

```python
B = np.array([[1,2,3], [3,2,1], [2,2,1]])

print(np.linalg.det(B))
```
